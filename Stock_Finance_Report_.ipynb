 DISCOUNTED CASH FLOW METHOD ~ VALUING A COMPANY 

Company cashflow :
1. Investment Activities
2. Financing Activities
3. Discounted Cashflow Models -> Operating Activities
    All the costs in an outflows related to the production and sales of a company good and services

    If utilizing income statements cannot see how much cash is generated
Net income contains noncash generated items when company doing sales and credits
Free cashflow after deducting capital expenditures for the operating cashflow, using free cashflow to estimate value of company

Weighted Average Cost of Capital in order to discount future free cashflow into the present moment to estimate the value of company

Forecast next five years income statement & cashflow statements to forecast free cashflow for the next five years of this company

https://codingandfun.com/calculating-weighted-average-cost-of-capital-wacc-with-python/


Weight Average Cost Capital -> Firm cost capital pay on average to all the security holders

WACC = Kd * (1 -Tc) * (D /D+E) + Ke * (E /D+E)

WACC = Weighted Average Cost of Capital
Kd = Cost of debt
Tc = Firm tax rate
Ke = Cost of equity
D / D + E = Proportion of debt in firm capital structure

Bond yield -> gl2 

Cost of Debt = Risk free rate + credit spread
Credit rating 3.8 good low risk low spread
Risky -> B- rating B3 / CCC rating bigger spread -> cost of debt is higher

Calculate credit spread based on interest coverage ratio

Interest Coverage Ratio : 8.5 means that company can cover interest that are going to pay 8 times or more
cost of debt = add spread to risk free rate

interest coverage ratio= net earnings (before interest & tax)/ interest expense

```
import pandas_datareader.data as web
import datetime
import requests
import pandas as pd

company='MSFT'


#IS = requests.get(f"https://financialmodelingprep.com/api/v3/income-statement/{company}?period=quarter&apikey={apikey}").json()

#print(IS)

def interest_coverage_and_RF(company):
  IS= requests.get(f'https://financialmodelingprep.com/api/v3/income-statement/{company}?apikey={demo}').json()
  EBIT= IS[0]['ebitda'] - IS[0]['depreciationAndAmortization'] #income statement recent year
  interest_expense = IS[0]['interestExpense']
  interest_coverage_ratio = EBIT / interest_expense

  #RF
  start = datetime.datetime(2019, 7, 10)
        
  end= datetime.datetime.today().strftime('%Y-%m-%d')
  #end = datetime.datetime(2020, 7, 10)
  Treasury = web.DataReader(['TB1YR'], 'fred', start, end) #1 year treasury rate
  RF = float(Treasury.iloc[-1]) #the last element of the list recent risk free rate
  RF = RF/100
  print(RF, interest_coverage_ratio)
  return [RF,interest_coverage_ratio]

interest_coverage_and_RF(company) 

interest = interest_coverage_and_RF(company) 
RF = interest[0]
interest_coverage_ratio = interest[1]

#Risk free ratio is almost zero, interest coverage ratio = 16 -> 16 times earnings to cover interest expenses

def cost_of_debt(company, RF,interest_coverage_ratio):
  if interest_coverage_ratio > 8.5:
    #Rating is AAA
    credit_spread = 0.0063
  if (interest_coverage_ratio > 6.5) & (interest_coverage_ratio <= 8.5):
    #Rating is AA
    credit_spread = 0.0078
  if (interest_coverage_ratio > 5.5) & (interest_coverage_ratio <=  6.5):
    #Rating is A+
    credit_spread = 0.0098
  if (interest_coverage_ratio > 4.25) & (interest_coverage_ratio <=  5.49):
    #Rating is A
    credit_spread = 0.0108
  if (interest_coverage_ratio > 3) & (interest_coverage_ratio <=  4.25):
    #Rating is A-
    credit_spread = 0.0122
  if (interest_coverage_ratio > 2.5) & (interest_coverage_ratio <=  3):
    #Rating is BBB
    credit_spread = 0.0156
  if (interest_coverage_ratio > 2.25) & (interest_coverage_ratio <=  2.5):
    #Rating is BB+
    credit_spread = 0.02
  if (interest_coverage_ratio > 2) & (interest_coverage_ratio <=  2.25):
    #Rating is BB
    credit_spread = 0.0240
  if (interest_coverage_ratio > 1.75) & (interest_coverage_ratio <=  2):
    #Rating is B+
    credit_spread = 0.0351
  if (interest_coverage_ratio > 1.5) & (interest_coverage_ratio <=  1.75):
    #Rating is B
    credit_spread = 0.0421
  if (interest_coverage_ratio > 1.25) & (interest_coverage_ratio <=  1.5):
    #Rating is B-
    credit_spread = 0.0515
  if (interest_coverage_ratio > 0.8) & (interest_coverage_ratio <=  1.25):
    #Rating is CCC
    credit_spread = 0.0820
  if (interest_coverage_ratio > 0.65) & (interest_coverage_ratio <=  0.8):
    #Rating is CC
    credit_spread = 0.0864
  if (interest_coverage_ratio > 0.2) & (interest_coverage_ratio <=  0.65):
    #Rating is C
    credit_spread = 0.1134
  if interest_coverage_ratio <=  0.2:
    #Rating is D
    credit_spread = 0.1512

    
  cost_of_debt = RF + credit_spread
  print(cost_of_debt)
  return cost_of_debt

kd = cost_of_debt(company,RF,interest_coverage_ratio)

kd
```

```
def costofequity(company):
  #RF
  start = datetime.datetime(2019, 7, 10)
  end= datetime.datetime.today().strftime('%Y-%m-%d')
  #end = datetime.datetime(2020, 7, 10)
  Treasury = web.DataReader(['TB1YR'], 'fred', start, end)
  RF = float(Treasury.iloc[-1])
  RF = RF/100
#Calculate market return with sp500 good proxy of risk of the market
#Capital Asset Pricing Model
#Beta
  beta = requests.get(f'https://financialmodelingprep.com/api/v3/company/profile/{company}?apikey={demo}')
  beta = beta.json()
  beta = float(beta['profile']['beta'])

  #Market Return
  start = datetime.datetime(2019, 7, 10)
  end= datetime.datetime.today().strftime('%Y-%m-%d')

  SP500 = web.DataReader(['sp500'], 'fred', start, end)
      #Drop all Not a number values using drop method.
  SP500.dropna(inplace = True)

  SP500yearlyreturn = (SP500['sp500'].iloc[-1]/ SP500['sp500'].iloc[-252])-1
    
  cost_of_equity = RF+(beta*(SP500yearlyreturn - RF))
  print(cost_of_equity)
  return cost_of_equity

ke = costofequity(company)
```

```
#tax & capital structure from balance sheet
#rax rate from financial ratio
#capital structure from balance sheet
#total debt & total stockholders equity
def wacc(company):
#effective tax rate
  FR = requests.get(f'https://financialmodelingprep.com/api/v3/ratios/{company}?apikey={demo}').json() #Financial Ratio
  ETR = FR[0]['effectiveTaxRate'] #Recent Effective Tax Rate

#capital structure (from balance statement)
  BS = requests.get(f'https://financialmodelingprep.com/api/v3/balance-sheet-statement/{company}?period=quarter&apikey={demo}').json()
  Debt_to = BS[0]['totalDebt'] / (BS[0]['totalDebt'] + BS[0]['totalStockholdersEquity'])  #Debt to Equity ratio / Leveraging Ratio
  equity_to = BS[0]['totalStockholdersEquity'] / (BS[0]['totalDebt'] + BS[0]['totalStockholdersEquity']) #Equity to Pricing Ratio
#ke = cost of equity
#wacc
  WACC = (kd*(1-ETR)*Debt_to) + (ke*equity_to)
  print(WACC,equity_to,Debt_to)
  return WACC
wacc_company = wacc(company)
print('wacc of ' + company + ' is ' + str((wacc_company*100))+'%')
```

FCFFn will be the last year of the forecasted free cash flows. Then, we need to come up with a perpetuity growth rate g. The perpetuity rate is the constant rate that a firm is expected to grow at forever.

We will assume 2% perpetuity rate since it is a common practice to be in line with the long-term inflation rate which is usually around 2%. Then, we discount the value using the WACC rate.

```
#Terminal value
LTGrowth = 0.02

Terminal_value = (CF_forecast['next_5_year']['FCF'] * (1+ LTGrowth)) /(wacc_company  - LTGrowth)

Terminal_value_Discounted = Terminal_value/(1+wacc_company)**4
Terminal_value_Discounted
```
